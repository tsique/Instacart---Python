# 02. Mixed -Type Data

```python
# Checking for mixed-type 

for col in df_ords.columns.tolist():
    weird = (df_ords[[col]].applymap(type) != df_ords[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_ords[weird]) > 0:
        print (col)
```

### Nothing was printed, so the statement is true and no mixed-type data has been found in the dataset

# 03. Missing Values

```python
# Checking for missing values in the dataset

df_ords.isnull().sum()
```

### In days_since_prior_order there are 206,209 missing values

```python
# Creating a new dataframe, df_nan, of the missing values 

df_nan = df_ords[df_ords['days_since_prior_order'].isnull() == True]
```

```python
# Checking df_nan was created properly

df_nan
```

## Addressing Missing Values

```python
# Creating a dataframe without the missing values

df_ords_clean = df_ords[df_ords['days_since_prior_order'].isnull() == False]
```

```python
df_ords_clean
```

## Creating a new dataframe we do not risk losing data through deleting and using a mean or median would not be accurate for later calculations.

# 04. Duplicate Values

```python
# Creating a subset of only rows that are duplicates in df_ords_clean

df_dups = df_ords_clean[df_ords_clean.duplicated()]
```
