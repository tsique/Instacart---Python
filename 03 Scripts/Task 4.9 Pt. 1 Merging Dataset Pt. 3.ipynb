```python
# Importing Libraries 

import pandas as pd 
import numpy as np 
import os 
```

```python
# Importing customers.csv 

df_cust = pd.read_csv (os.path.join(path, '02 Data', 'Original Data', 'customers.csv'), index_col = False)
```

# 01. Data Wrangling

```python
# Output check

df_cust
```

### Renaming columnns

```python
# Renaming STATE

df_cust.rename(columns = {'STATE' : 'state'}, inplace = True)
```

```python
# Renaming Gender

df_cust.rename(columns = {'Gender' : 'gender'}, inplace = True)
```

```python
# Renaming Age

df_cust.rename(columns = {'Age' : 'age'}, inplace = True)
```

```python
# Renaming n_dependants

df_cust.rename(columns = {'n_dependants' : 'number_dependents'}, inplace = True)
```

```python
# Checking output 

df_cust
```

### Dropped columns 

```python
# Dropping First Name, Surnam, date_joined
# Creating new dataframe in case columns are wanted later 

df_cust_2 = df_cust.drop(columns = ['First Name', 'Surnam', 'date_joined'])
```

```python
# Checking output 

df_cust_2
```

# 02. Quality and Consistent Check

### Missing Values

```python
# Checking for misssing values

df_cust_2.isnull().sum()
```

No missing values are detected

### Duplicates

```python
# Checking for duplicates

df_dups = df_cust_2[df_cust_2.duplicated()]
```

```python
# Checking output 

df_dups
```

No duplicated values were founded

### Mixed type data

```python
# Checking for mixed-type 

for col in df_cust_2.columns.tolist():
    weird = (df_cust_2[[col]].applymap(type) != df_cust_2[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_cust_2[weird]) > 0:
        print (col)
```

Nothing was printed so the statement is true and no mixedup data was found
